# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
# Count of Female Buyers.

Female_count=cars_data[cars_data['Buyer Gender']=='Female'].count()
Female_count=Female_count['Buyer Gender']
Female_count
```

```python
# Count of Male Buyers.

Male_count=cars_data[cars_data['Buyer Gender']=='Male'].count()
Male_count=Male_count['Buyer Gender']
Male_count
```

```python
counts=[Female_count,Male_count]
counts
```

```python
# bar graph of male vs female buyers participated in the sales:

plt.bar(x_ticks,counts,width = 0.4,color=['orange','blue'] )
x_ticks=['Female','Male']
x=['Female','Male']
plt.ylabel('Number of buyers',color='green')
plt.ylim(0,8000)
plt.title('Male vs Female Buyers participated in the sales ',color='red',fontsize=12)

# Annotations:
for i,j in zip(x,counts):
    plt.annotate(str(j),(i,j+100),color='black',size=10)
    
plt.show()
```
